```python
import nltk
nltk.download('punkt')  # Descarga los modelos necesarios para dividir el texto en oraciones (sent_tokenize) y palabras (word_tokenize)
nltk.download('averaged_perceptron_tagger')  # Descarga el modelo para etiquetado gramatical (pos_tag)
nltk.download('maxent_ne_chunker')  # Descarga el modelo para análisis de entidades nombradas (ne_chunk)
nltk.download('words')  # Descarga la lista de palabras en inglés
nltk.download('stopwords')  # Descarga la lista de palabras de detención (stopwords) en inglés

# Función para procesar el texto
def process_text(text):
    sentences = sent_tokenize(text)  # Divide el texto en oraciones
    print("Oraciones:", sentences)

    tokens = word_tokenize(text)  # Divide el texto en palabras (tokens)
    print("Tokens:", tokens)

    tagged = pos_tag(tokens)  # Asigna etiquetas gramaticales a cada palabra
    print("Etiquetas gramaticales:", tagged)

    named_entities = ne_chunk(tagged)  # Realiza el análisis de entidades nombradas
    print("Entidades nombradas:", named_entities)

    stop_words = set(stopwords.words('english'))  # Crea un conjunto de palabras de detención en inglés
    filtered_words = [token for token in tokens if token.lower() not in stop_words]  # Filtra las palabras de detención del texto
    print("Palabras filtradas:", filtered_words)

# Nombre del archivo de texto a leer
nombre_archivo = "ejemplo.txt"

# Leer el contenido del archivo
with open(nombre_archivo, 'r') as archivo:
    texto = archivo.read()

# Procesar el texto
process_text(texto)
```
